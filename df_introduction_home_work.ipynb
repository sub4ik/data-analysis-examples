```python
df = pd.read_csv('data/Minjust2018.csv', header = None, usecols = [4,14,16,19,27,28,33])
df.fillna(' ', inplace = True)
df.rename(columns = {4:"name",14:'city',16:'street',19:'phone_number',27:"director",28:"activity",33:"founders"}, inplace = True)
```

```python
#Найти все компании, которые находятся на вашей улице
df.loc[(df['city'] == 'Бишкек') & (df['street'] == 'Гагарина'), ['name','city','street']]
```

```python
#Найти все компании, которые принадлежат вашим тезкам
df.loc[(df['founders'].str.contains('Субан')), ['name','founders']]
```

```python
#Найти все компании, которые находятся на вашей улице и принадлежат вашим тезкам
df.loc[((df['city'] == 'Бишкек') & (df['street'] == 'Гагарина')) & (df['founders'].str.contains('Чынара')), :]
```

```python
#Найти все компании, которые занимаются строительством и в своем телефоне имеют цифры 432 подряд
df.loc[((df['phone_number'].str.contains('432'))|
       (df['phone_number'].str.contains('4-3-2'))|
       (df['phone_number'].str.contains('43-2'))|
       (df['phone_number'].str.contains('4-32'))|
       (df['phone_number'].str.contains('4 32'))|
       (df['phone_number'].str.contains('43 2')))&(df['activity'].str.contains('строит'))]
#df['phone_number'][140]
```
